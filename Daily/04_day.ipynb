```python
# Task class(oop)
import datetime

class Task:
    def __init__(self,task_id, title, description, due_date, is_done):
        self.task_id = int(task_id)
        self.title = title
        self.description = description
        self.due_date = datetime.date.fromisoformat(due_date)
        self.is_done = bool(is_done)

    #   ---METHODS---

    def to_dict(self):
        '''converts input into dict'''
        return {
            "task_id":self.task_id,
            "title":self.title,
            "description": self.description,
            "due_date": self.due_date.isoformat(), # isoformat will convert it into string because json dont save datetime object.
            "is_done": self.is_done
        }
    
    def mark_done(self):
        '''mark task complete'''
        self.is_done = True

    def is_due(self):
        '''checks due date'''
        today = datetime.date.today()
        return (today == self.due_date) and (not self.is_done)

    def is_overdue(self):
        '''checks if its overdue'''
        today = datetime.date.today()
        return (today > self.due_date) and (not self.is_done)
    

    # still continue
```
